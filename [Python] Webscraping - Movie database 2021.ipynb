```python
import requests
from bs4 import BeautifulSoup as soup
page = requests.get("https://movieweb.com/movies/2021/")
page.status_code
```

```python
scraped = soup(page.content, 'lxml')
base = 'https://movieweb.com'
links = []
for movie in scraped.find_all('a',class_='movie-wrap'):
        links.append(base+movie['href'])
len(links) #77 
```

```python
movies = []
for link in links:
    try:
        page = requests.get(link)
        scraped = soup(page.content, 'lxml')

        Site = link
        Release = Title = Genre = Runtime = Director = Writers = Cast = Synopsis = ''
    
        Release = scraped.find_all('dd', itemprop='datePublished')[0].text.strip()
        Title = scraped.select('.movie-title')[0].text.strip().replace('(2021)','')
        Synopsis = scraped.find_all('p', itemprop='description')[0].text.strip()
    
        Genre = scraped.find_all('span', itemprop='genre')[0].text.strip()
        Director = scraped.find_all('ul', itemprop='director')[0].text.strip().replace('\n\n\n',', ')
        Writers = scraped.find_all('ul', itemprop='author')[0].text.strip().replace('\n\n\n',', ')
        Cast = scraped.find_all('ul', itemprop='actors')[0].text.strip().replace('\n\n\n',', ')
        Runtime = scraped.select('.movie-runtime')[0].text.strip()

        movies.append((
            Release, Title, Genre, Runtime, Director, Writers, Cast, Synopsis, Site  
        )) 
    except IndexError:
        movies.append((
            Release, Title, Genre, Runtime, Director, Writers, Cast, Synopsis, Site       
        ))

len(movies)
```

```python
import pandas as pd
pd.set_option('display.max_rows', None)
movies_df = pd.DataFrame(movies,
                     columns=['Release Date','Title','Genre','Runtime',
                              'Director','Writers','Cast','Synopsis','Link'])
movies_df
```
